# Project 1 - Library Book Management System

```python
from tkinter import *
import sqlite3 as sql
import tkinter.ttk as ttk
import tkinter.messagebox as tkMessageBox
```

```python
root = Tk()
root.title('Library Book Management System')

#*************************************************//METHOD//******************************************************************

#creation of database
def Database():
    global conn, cursor
    conn = sql.connect('books.db')
    cursor = conn.cursor()
    cursor.execute('CREATE TABLE IF NOT EXISTS book_mangement(book_id INTEGER PRIMARY KEY AUTOINCREMENT NOT NULL,book TEXT, author TEXT,year TEXt, isbn TEXT,issue TEXT )')



# create display function
def Display():
    tree.delete(*tree.get_children())
    Database()
    cursor.execute("SELECT * FROM book_mangement ORDER by book ASC")
    fetch = cursor.fetchall()
    for data in fetch:
        tree.insert('','end',value=(data[1],data[2],data[3],data[4]))
    cursor.close()
    txt_result.config(text="Successfully Done", fg = 'black')



# create search function
def Search():
    if BOOK.get() == "" and AUTHOR.get() == "" and YEAR.get() == "":
        txt_result.config(text = "Please enter")
    else:
        tree.delete(*tree.get_children())
        Database()
        cursor.execute("SELECT * FROM book_mangement where book = ? or author = ? or year = ?",(str(BOOK.get().title()),str(AUTHOR.get().title()),str(YEAR.get())))
        fetch = cursor.fetchall()
        if len(fetch) == 0:
            clear_entry()
            cursor.close()
            txt_result.config(text="Book Not Available", fg = 'red')
        else:
            for data in fetch:
                tree.insert('','end',value=(data[1],data[2],data[3],data[4]))
            clear_entry()

            cursor.close()
            txt_result.config(text="Successfully fetched", fg = 'green')



# create add function
def Add():
    if BOOK.get() == "" or AUTHOR.get() == "" or YEAR.get() == "":
        txt_result.config(text = "Please enter required entry")
    else:
        Database()
        val = str(ISBN.get())
        if len(val) == 0:
            val = ''
        cursor.execute("INSERT INTO book_mangement(book, author, year, isbn, issue) VALUES (?,?,?,?,?)",(str(BOOK.get().title()),str(AUTHOR.get().title()),str(YEAR.get()),val,''))
        conn.commit()
        clear_entry()
        cursor.close()
        conn.close()
        txt_result.config(text='Recorded',fg = "green")

        

#create issue function
def Issue():
    if BOOK.get() == "" or AUTHOR.get() == "" or YEAR.get() == "":
        txt_result.config(text = "Please enter required entry")
    else:
        Database()
        cursor.execute("SELECT * FROM book_mangement where book = ? and author = ? and year = ?",(str(BOOK.get().title()),str(AUTHOR.get().title()),str(YEAR.get())))
        fetch = cursor.fetchall()
        for data in fetch:
            if data[5] != '':
                cursor.close()
                txt_result.config(text="Book Not Available", fg = 'red')
            else:
                cursor.execute('UPDATE book_mangement SET issue = (?) where book = ? and author = ? and year = ?',('issued',str(BOOK.get().title()),str(AUTHOR.get().title()),str(YEAR.get())))
                fetch = cursor.fetchall() 
                conn.commit()
                clear_entry()
                cursor.close()
                conn.close()
                txt_result.config(text='Issued Succesfully',fg = "green")


# create Delete fuction
def Delete():
    if BOOK.get() == "" or AUTHOR.get() == "" or YEAR.get() == "":
        txt_result.config(text = "Please enter required entry",fg = 'blue')
    else:
        Database()
        cursor.execute("SELECT * FROM book_mangement where book = ? and author = ? and year = ?",(str(BOOK.get().title()),str(AUTHOR.get().title()),str(YEAR.get())))
        fetch = cursor.fetchall()
        if len(fetch) == 0:
            clear_entry()
            cursor.close()
            txt_result.config(text="Incorrect value", fg = 'red')
        else:
            result = tkMessageBox.askquestion('Do you want to Delete? (y/n)', icon = 'warning')
            if result  == 'yes':
                Database()
                cursor.execute('DELETE FROM book_mangement where book = ? and author = ? and year = ?',(str(BOOK.get().title()),str(AUTHOR.get().title()),str(YEAR.get())))
                conn.commit()
                clear_entry()
                conn.close()
                Display()
    
    

# create Exit function
def Exit():
    result = tkMessageBox.askquestion('Do you want to exit? (y/n)', icon = 'warning')
    if result  == 'yes':
        root.destroy()
        exit()

# crete a entry box clearing function
def clear_entry():
    BOOK.set("")
    AUTHOR.set("")
    YEAR.set("")
    ISBN.set("")
    
        
#=======================================================//VARIABLE//===============================================        
BOOK = StringVar()
AUTHOR = StringVar()
YEAR = StringVar()
ISBN = StringVar()


#===================================================//FRAME LAYOUT//==========================================================

Top = Frame(root, width = 100, height = 50, bd = 8, relief = 'raise', bg = 'blue')
Top.pack(side=TOP)

Left = Frame(root, width = 100, height = 500, bg = 'red', bd = 8, relief = 'raise')
Left.pack(side=LEFT)

Right = Frame(root, width = 100, height = 500, bg = 'yellow', bd = 8, relief = 'raise')
Right.pack(side=RIGHT)

Forms = Frame(Left, width = 200, height = 100, bd = 8, relief = 'raise')
Forms.pack(side=TOP)

Buttons = Frame(Left, width = 400, height = 100, bd = 8, relief = 'raise')
Buttons.pack(side=BOTTOM)

Right_scroll = Frame(Right, width = 400 , height = 250, bd = 8, relief = 'flat')
Right_scroll.pack(side=RIGHT)

#================================================//LABEL//================================================================

txt_title = Label(Top, width = 60 ,font= ('airel', 20), text = '...Library Book Management System...')
txt_title.pack()

empty = Label(root, height = 40)
empty.pack()
#txt_title = Label(Top, Width = 900, text = 'Library Book Management System')
#txt_title.pack()
txt_booktitle = Label(Forms, text = 'Booktitle:', font = ('airel', 15), bd = 15)
txt_booktitle.grid(row = 0, stick = 'e')
txt_author = Label(Forms, text = 'Author:', font = ('airel', 15), bd = 15)
txt_author.grid(row = 1, stick = 'e')
txt_year = Label(Forms, text = 'Year:', font = ('airel', 15), bd = 15)
txt_year.grid(row = 2, stick = 'e')
txt_isbn = Label(Forms, text = 'ISBN:', font = ('airel', 15), bd = 15)
txt_isbn.grid(row = 3, stick = 'e')
txt_result = Label(Buttons)
txt_result.pack(side = TOP)



#=====================================================//ENTRY//==============================================================

booktitle = Entry(Forms, textvariable = BOOK, width = 30)
booktitle.grid(row = 0,column = 1)
author = Entry(Forms, textvariable = AUTHOR, width = 30)
author.grid(row = 1,column = 1)

year = Entry(Forms, textvariable = YEAR, width = 30)
year.grid(row = 2,column = 1)
isbn = Entry(Forms, textvariable = ISBN, width = 30)
isbn.grid(row = 3,column = 1)

#=======================================================//BUTTONS//===============================================================

btn_display = Button(Buttons, width=10, text='Display', command = Display)
btn_display.pack(side=LEFT)
btn_search = Button(Buttons, width=10, text='Search', command = Search)
btn_search.pack(side=LEFT)
btn_add = Button(Buttons, width=10, text='Add', command = Add)
btn_add.pack(side=LEFT)
btn_issue = Button(Buttons, width=10, text='Issue', command = Issue)
btn_issue.pack(side=LEFT)
btn_delete = Button(Buttons, width=10, text='Delete', command = Delete)
btn_delete.pack(side=LEFT)
btn_exit = Button(Buttons, width=10, text='EXIT', command = Exit)
btn_exit.pack(side=LEFT)

 
#========================================================//SCROLLBAR//==============================================================
tree = ttk.Treeview(Right_scroll, columns = ('Booktitle', 'Author', 'Year', 'ISBN','Status'))

scrollbar_y = Scrollbar(Right_scroll, orient = VERTICAL)
scrollbar_x = Scrollbar(Right_scroll, orient = HORIZONTAL)

scrollbar_y.config()
scrollbar_y.pack(side=RIGHT, fill = Y)
scrollbar_x.config()
scrollbar_x.pack(side = BOTTOM, fill = X)


tree.column('#0', width = 0, minwidth = 0)
tree.column('#1', width = 100, minwidth = 0, anchor = W)
tree.column('#2', width = 150, minwidth = 0, anchor = W)
tree.column('#3', width = 70, minwidth = 0, anchor = W)
tree.column('#4', width = 70, minwidth = 0, anchor = W)
tree.column('#5', width = 0, minwidth = 0, anchor = W)

tree.heading('#0')
tree.heading('Booktitle',text = 'Booktitle', anchor = W)
tree.heading('Author', text = 'Author', anchor = W)
tree.heading('Year', text = 'Year', anchor = W)
tree.heading('ISBN', text = 'ISBN', anchor = W)
tree.heading('Status', text = 'Status', anchor = W)

#tree.insert(parent='',index = 'end',iid=0, text = '', values= ('hbvfh','hfifhw','bhfiu','2625'))

tree.pack()



#====================================================//MAIN//=================================================================


if __name__ =='__main__':
    root.mainloop()
```
